\begin{align*}
		\mathbb{E}[X] &=  \sum_{k\in \{0,1\}}kP(X=k) \\
		&= 0\cdot P(X=0) + 1 \cdot P(X=1) \\
		&= P(X=1) = p \\
		Var(X) &= \mathbb{E}[X^2] - (\mathbb{E}[X])^2 \\
		&= \sum_{k\in \{0,1\}}k^2P(X=k) - p^2 \\
		&= p - p^2 = p(1-p) \\
		\mathbb{E}[e^{itX}] &= \sum_{k\in \{0,1\}}e^{itk}P(X=k) \\
		&= P(X=0) + e^{it}P(X=1) \\
		&= 1 - p + e^{it}p
	\end{align*}

# Q2.**Gaussian sums**
Let $X,Y$ be two gaussian with mean $\mu$ and variance $\sigma^2$

Since $X$ and $Y$ are independent, the joint probability density is the product of the pdfs of $X$ and $Y$. And so we have :
\begin{align*}
	\mathbb{E}[e^{it(X+Y)}] &= \int_{-\infty}^{\infty}\int_{-\infty}^{\infty} e^{it(x+y)} \frac{1}{\sqrt{2\pi \sigma^2}}e^{\frac{-(x-\mu)^2}{2\sigma^2}} \cdot \frac{1}{\sqrt{2\pi \sigma^2}}e^{\frac{-(y-\mu)^2}{2\sigma^2}} dx dy \\
	&=  \int_{-\infty}^{\infty}\int_{-\infty}^{\infty} e^{itx}e^{ity} \frac{1}{\sqrt{2\pi \sigma^2}}e^{\frac{-(x-\mu)^2}{2\sigma^2}} \cdot \frac{1}{\sqrt{2\pi \sigma^2}}e^{\frac{-(y-\mu)^2}{2\sigma^2}} dx dy \mbox{ because of the independence we get}\\
	&= \int_{-\infty}^{\infty} e^{itx} \frac{1}{\sqrt{2\pi \sigma^2}}e^{\frac{-(x-\mu)^2}{2\sigma^2}}dx \cdot \int_{-\infty}^{\infty} e^{ity} \frac{1}{\sqrt{2\pi \sigma^2}}e^{\frac{-(y-\mu)^2}{2\sigma^2}} dy \\
	&= e^{\frac{-t^2\sigma^2}{2}}e^{\mu it} \cdot e^{\frac{-t^2\sigma^2}{2}}e^{\mu it} \\
	&= e^{\frac{-t^2 (2\cdot \sigma^2)}{2}}e^{2\mu it} \\
	&= e^{-t^2 \cdot \sigma^2}e^{2\mu it}
\end{align*}
We get another gaussian random variables with parameters $\mu' = 2\mu$ and $\sigma'^2 = 2\sigma^2$.

To generalize to n sumes of i.i.d Gaussian, let's consider $\{X_i\}_{i=1}^{n}$ Gaussians with parameters $\{(\mu_i , \sigma^2_i)\}_{i=1}^{n}$. We would then have, by using the characteristic function :

\begin{align*}
\mathbb{E}[e^{it(X_1 + X_2 +\dots + X_n)}] &= \int_{-\infty}^{\infty}\int_{-\infty}^{\infty}…\int_{-\infty}^{\infty}e^{it(x_1 + x_2 + \dots + x_n)}\prod_{i=1}^{n}\frac{1}{\sqrt{2\pi\sigma^2_i}}e^{-\frac{(x - \mu_i)}{2\sigma_{i}^2}}dx_1dx_2\dots dx_n \\
&= \int_{-\infty}^{\infty}\int_{-\infty}^{\infty}…\int_{-\infty}^{\infty}\prod_{i=1}^{n}e^{itx_i}\frac{1}{\sqrt{2\pi\sigma^2_i}}e^{-\frac{(x - \mu_i)}{2\sigma_{i}^2}}dx_1dx_2\dots dx_n \\
&= \prod_{i=1}^{n} (\int_{-\infty}^{\infty} e^{itx_i}\frac{1}{\sqrt{2\pi\sigma^2_i}}e^{-\frac{(x - \mu_i)}{2\sigma_{i}^2}}dx_i) \\
&= \prod_{i=1}^{n} \mathbb{E}[e^{itX_i}] = \prod_{i=1}^{n} e^{-\frac{t^2\sigma_i^2}{2}}e^{i\mu_it} \\
&= e^{-\frac{t^2\sum_{i=1}^{n}\sigma_i^2}{2}}e^{it\sum_{i=1}^{n}\mu_i}
\end{align*}
*Conclusion:* The sum of $n$ iid Gaussian is also a Gaussian with parameter $(\sum_{i=1}^{n}\mu_i,\sum_{i=1}^{n}\sigma_i^2)$

# Q4.**Histogram**

```python
def myhistogram(v,a,b,n):
    """
    Returns the histogram of the vector v in the interval [a,b] with n bins, spread uniformly
    Args :
        - v [list]: an array of the data that we want to build an histogram
        - a [float]: the lower bound of the histogram
        - b [float]: the upper bound of the histogram
        - n [int]: the bin count
    Return :
        - myhist_count [list[int]] : the bin count in each of the partition
        - partition [list[float]] : the subdivision for the bin count
    """
    assert b > a
    step = (b - a)/(n)
    count = 0
    myhist_count = []

    start = a
    next_step = a + step
    partition = [a + i * (step) for i in range(n+1)]

    for low, up in zip(partition[:-1], partition[1:]):
        for element in v :
            if low <= element and element < up:
                count +=1
        myhist_count.append(count)
        count = 0    
    
    return myhist_count, partition
```

```python
import numpy as np
n = 100
bins = 100
X = np.random.randn(n)
vec = myhistogram(X, np.min(X), np.max(X), bins)
npvec = np.histogram(X, bins)
```

# Q5.**Non-uniform variates**
Use the transformation method to construct random number generators for the
following probability distributions and test them with large-enough samples by plotting the sample histogram
(properly normalised) with the corresponding theoretical distribution.

- *Random choice $[1,2,\dots,N]$ with probability $[p_1,p_2,\dots , p_N]$*.
First of all we have the pdf of the random choice as $p(x) = \sum_{k=1}^{N} \mathbb{I}_{\{x=k\}}p_k$
In that case the cdf is :
$$F(x) = \sum_{k=1}^{N} \mathbb{I}_{\{k \le x\}}p_k$$ and we have 
$$
x =
\begin{cases}
	1 & \text{if } 0 \le u < p_1 \\
	2 & \text{if } p_1 \le u < p_1  + p_2\\
	3 & \text{if } p_1 + p_2\le u < p_1  + p_2 + p_3\\ 
	 \vdots \\
	N-1 & \text{if } p_1 + p_2 + \dots + p_{N-2}\le u < p_1 + p_2 + \dots + p_{N-2} + p_{N-1} \\
	N & \text{if } p_1 + p_2 + \dots + p_{N-1}\le u < 1
\end{cases}
$$

- *Uniform over $[-1,1]$*
The pdf is $f(x) = \frac{1}{2}\mathbb{1}_{[-1,1]}$. The cumulative distribution function is  :
	\begin{align*}
		F(x) &= \int_{-\infty}^{x}\frac{1}{2}\mathbb{I}_{[-1,1]} dx \\
		&= \int_{-1}^{x}\frac{1}{2}\mathbb{I}_{[-1,1]} dt \\
		&= \frac{1}{2}\left[t\right]_{-1}^{x} \\
		&= \frac{1}{2}(x+1) = u \mbox{ we then have : }\\
		2u &= x +1 \\
		x &= 2u - 1  
	\end{align*}

- *Exponential with parameter $\lambda$*
The pdf is $f(x) = \lambda e^{-\lambda x}$. The cumulative distribution function is  :
	\begin{align*}
		F(x) &= \int_{0}^{x}\lambda e^{-\lambda t} dt \\
		&= \left[-e^{-\lambda t}\right]_{0}^{x} \\
		&= 1 - e^{-\lambda x} = u \mbox{ we then have : } \\
		1 - u &= e^{-\lambda x} \\
		\ln(1 - u) &= -\lambda x \\
		x &= \frac{-1}{\lambda}\ln(1 - u)
	\end{align*}

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def custom_random_choice(u, params=[]):
    count = 0
    if params[0]:
        cum_sum = np.cumsum(params)
        return np.count_nonzero(cum_sum[cum_sum <= u]) + 1

def custom_uniform(u):
    return 2*u - 1

def custom_exponential(u, lambda_param) :
    return -1/lambda_param * (math.log(1 - u))

cur_samples = 10**4
params = np.array([0.125, 0.5, 0.25, 0.125])
data_0 = np.array([custom_random_choice(np.random.rand(), params) for i in range(cur_samples)])
data_1 = np.array([custom_uniform(np.random.rand()) for i in range(cur_samples)])
data_2 = np.array([custom_exponential(np.random.rand(),2) for i in range(cur_samples)])

plt.hist(data_1, 100, density=True)
plt.plot([-1,1], [0.5,0.5])
plt.title("Custom uniform versus Real theortical uniform")
plt.show()
```

# Q7.**Monte Carlo estimation of $\pi$**
Choose a point $(x,y)$ at random in the square.
$$
S = \{(x,y): x \in [-1,1], y \in [-1,1]\}
$$
The probability that the point lies in the circle :
$$
C = \{(x,y): x^2 + y^2 = 1\}
$$
is equal to $\pi / 4$

```python
import numpy as np
tries = 10**3 #this taks too much time compared to the GPU of course

u = np.random.rand(tries)
v = np.random.rand(tries)
X = custom_uniform(u)
Y = custom_uniform(v)
mean_est = np.zeros(tries)
mean_est_err = np.zeros(tries)

true_count = 0

for idx,x,y in zip(range(tries), X, Y):
    norm = np.square(x) + np.square(y)
    if  norm <= 1 :
        true_count += 1
    mean_est[idx] = 4 * true_count/(idx+1)
    sec_mom = mean_est[idx]**2 / (idx+1)
    mean_est_err[idx] = np.sqrt(np.abs(sec_mom - mean_est[idx]**2)) / np.sqrt(idx+1.0) 

# Plot estimator with error bar
plt.errorbar(range(tries), mean_est, mean_est_err, errorevery=10, ecolor='green')

# Show theoretical value
plt.plot(range(tries), np.zeros(tries)+np.pi, 'k--')
plt.title("Pi approximation by Monte Carlo methods")
plt.show()
```

# Q8.**Two-state Markov chain**

```python
import random
import numpy as np

class State:
    def __init__(self, desc=0, param=0.5):
        self.desc = desc
        self.param = param

    def __str__(self):
        return str(self.desc)

    def __eq__(self, other):
        return isinstance(other, State) and self.desc == other.desc

    def next_step(self):
        res = np.random.binomial(1, self.param)
        return res == 0 # should i switch?


class Markov_Chain:
    def __init__(self, num_state=2, alphas=[], N_trials=1000):
        self.N_trials = N_trials
        self.states = []
        self.recorder = []
        self.result = []
        self.alphas = alphas
        for i, alpha in enumerate(self.alphas):
            self.states.append(State(desc=i, param=alpha))

    def random_walk(self):
        self.recorder.append(np.random.choice(a=self.states))
        for trial in range(self.N_trials-1):
            if self.recorder[-1].next_step() :
                # stay in the actual state
                self.recorder.append(self.recorder[-1])
            else :
                # Choose the other one
                other = [s for s in self.states if s != self.recorder[-1]]
                self.recorder.append(other[0])


        self.result = [record.desc for record in self.recorder]

    def ratio_result(self):
        count = 0
        for state_count in self.recorder:
            if state_count == self.states[0]:
                count += 1
        print("Predicted Ratio : ", count/self.N_trials, (self.N_trials-count)/(self.N_trials))
        print("Theoritical Ratio : ", self.alphas[1]/(self.alphas[1] + self.alphas[0]) , self.alphas[0]/(self.alphas[1] + self.alphas[0]))
```

Generate a connected graph, say with 10 or more states, and write a program that simulates a trajectory of the unbiased random walk on that graph. Plot a trajectory up to 1000 times steps and use it to verify that the fraction of time a node $i$ is visited is proportional to its degree $k_i$, so that nodes with higher degree are most oftern visited. The Python package networkx is useful for this question.

```python
import networkx as nx
import random

from collections import Counter
```

```python
num_state = 15

G = nx.erdos_renyi_graph(20,p=0.2)
nx.draw_random(G, with_labels=True)
```

```python
traj = []
length = 1000
start = random.randint(0,num_state)
traj.append(start)

for _ in range(length):
    curr_neighbor = G.neighbors(traj[-1])
    next_step = random.choice(list(curr_neighbor))
    traj.append(next_step)


counts = Counter(traj)
x = list(counts.keys())
y = list(counts.values())

plt.bar(x, y)
plt.xlabel("Value")
plt.ylabel("Occurrences")
plt.title("Count of each integer")
plt.show()
```

```python
for i in range(num_state):
    print(f"Node #{i} has {len(list(G.neighbors(i)))} neighbors \n")
```
